```python
# Difference between seaborn and matplotlib
"""
Matplotlib settings are difficult to figure out. Seaborn comes with different customization and high level interface.

Matplotlib dosnt serve well when it comes to dealing with dataframes, While seaborn works well with dataframes.
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
sns.relplot(x='passengers', y = 'month',data = a)
```

```python
a = sns.load_dataset("flights")
sns.relplot(x='passengers', y = 'month',hue = 'year',data = a)
```

```python
a = sns.load_dataset("flights")
sns.relplot(x='month', y = 'passengers',hue = 'year',data = a)
```

```python
a = sns.load_dataset("tips")
sns.relplot(x='time', y = 'tip',data = a,kind = 'line')
```

```python
sns.catplot(x = 'day',y='total_bill',data = a)
```

```python
sns.catplot(x = 'day',y='total_bill',data = a,kind = 'violin')
```

```python
sns.catplot(x = 'day',y='total_bill',data = a,kind = 'boxen')
```

```python
from scipy import stats
c = np.random.normal(loc = 5, size = 100, scale =2)
```

```python
sns.distplot(c)
```

```python
a = sns.load_dataset('iris')
b = sns.FacetGrid(a, col = 'species')
b.map(plt.hist, 'sepal_length')
```

```python
a = sns.load_dataset('flights')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style = 'darkgrid')
a = sns.load_dataset('flights')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='white',color_codes = True)
a = sns.load_dataset('tips')
sns.boxplot(x = 'day', y = 'total_bill',data =a)
```

```python
sns.set(style='white',color_codes = True)
a = sns.load_dataset('tips')
sns.boxplot(x = 'day', y = 'total_bill',data =a)
sns.despine(offset = 10, trim = True )
```
